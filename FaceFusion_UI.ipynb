## BASIC GUIDE:
1. Choose a runtime type with Runtime -> Change runtime type (CPU is slower, any GPUs like the T4 free daily GPU are faster), Run the Install FaceFusion cell and wait until it finishes.
2. (OPTIONAL) Run the Mount Google Drive cell, by default it will make your Google Drive FF_Colab/Outputs and use it as the output path.
3. **(ONLY FOR WHO MOUNTED DRIVE)** Now you got 2 ways to upload the inputs:

**For Google Chrome Users:**
1. Run the Mount Google Drive Cell

2. Run the 'Upload Target Vid/Pic & Source Pic' cell and upload both of them (either do ctrl and select both of them or just select one, then run it again, and select the other)

3. Wait until the Target and Source files are loaded, and then in the Run Face Fusion cell, in Target_Video insert the name of the Video or the Target Photo (also insert the extension of your Source & Target, e.g. source.jpeg, target.mp4, target.png, output.mp4, output.jpg), for Source_Image put the name of the photo Source, and for Output insert the name you want to give to the final output (with the same extension of the Target).

**For NON Google Chrome Users:**
1. Run the Mount Google Drive Cell

2. Go to [Your Google Drive](https://drive.google.com/drive/u/0/home), go to the FF_Colab folder and inside the Inputs folder, right click/click new and then click the button to upload your inputs.

3. Wait until the Target and Source files are loaded, and then in the Run Face Fusion cell, in Target_Video insert the name of the Video or the Target Photo (also insert the extension of your Source & Target, e.g. source.jpeg, target.mp4, target.png, output.mp4, output.jpg), for Source_Image put the name of the photo Source, and for Output insert the name you want to give to the final output (with the same extension of the Target).


2. Select which type of Tunnel you wanna use, so if the API of one of them is down, you can use the other one:

 A) Gradio: Select it in Tunnel, Run the Run cell and click the Gradio Public Link Url.

 B) Ngrok: Select it in Tunnel, get the Ngrok Tunnel Authtoken here: https://dashboard.ngrok.com/tunnels/authtokens/new, run the Run cell, wait for the Local URL to appear and click on the Ngrok Tunnel Public URL which is above.

  C) Cloudflare: Select it in Tunnel, run the Run cell, wait for the Local URL to appear and click on the Cloudflare Tunnel Public URL which is above.

  D) LocalTunnel: Select it in Tunnel, run the Run cell, wait for the Local URL to appear, copy the LocalTunnel Password displayed under the public link and paste it in Tunnel Password of the LocalTunnel Tunnel Public URL which is above.

4. For Source select the photo of the person taken as a reference point.
5. For Target, select the photo/video of the person whose face you want to change.
6. **(OPTIONAL)** You can decide to activate the Frame Processors face_enhancer for better results in some cases.
7. Under Face Recognition you can choose Many to change all Target faces or Reference to change only certain reference faces.
8. **(OPTIONAL)** With Reference Face Distance you can choose the radius distance of the reference faces that the program will change.
9. **(OPTIONAL)** Under Reference Analyzer Direction you can choose the radius that identifies the faces to change, from left to right or vice versa.
10. **(OPTIONAL)** Using Face Analyzer Age you can decide whether to change the faces of only people of a certain age.
11. **(OPTIONAL)** Through Face Analyzer Gender you can decide whether to change the faces of only people of a certain sexual gender.
12. **(OPTIONAL)** Under Options you can select Skip Download to automatically download the output.
13. Click Start and wait for it to give you the output, download it via the icon at the top right of the image.
